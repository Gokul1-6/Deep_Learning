```python
from transformers import pipeline
sentiment_pipeline = pipeline("sentiment-analysis")
sentiment_pipeline("today is a happiest day for me")
```

```python
classifier = pipeline("zero-shot-classification")
classifier(
    "i am going to play cricket",
    candidate_labels=["education","politics","sports"]
)
```

```python
generator = pipeline("text-generation")
generator("today lunch is")
```

```python
generator = pipeline("text-generation",model="SupraLabs/Supra-50M-Instruct")
generator(
    "i want to play a",
    max_length=30,
    num_return_sequences=2,
)
```

```python
generator = pipeline("fill-mask",model="DeepChem/MoLFormer-c3-1.1B")
generator(
    "i want to play a ",
    max_length=30,

)
```
